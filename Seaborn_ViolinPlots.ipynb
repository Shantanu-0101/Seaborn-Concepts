# ViolinPlot

```python
import seaborn as sns
import numpy as np
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
sns.violinplot(tips['total_bill'])
```

```python
sns.catplot(x='day', y='total_bill', kind='violin', data=tips)
```

```python
sns.catplot(x='day', y='total_bill', kind='violin', hue='sex', data=tips)
```

```python
sns.catplot(x='day', y='total_bill', kind='violin', hue='sex',split='True', data=tips)
```

# Barplot:

```python
sns.catplot(x='smoker', y='total_bill', kind='bar', data=tips)
```

```python
sns.catplot(x='smoker', y='total_bill', hue='sex', kind='bar', data=tips)
```

```python
sns.catplot(x='smoker', y='total_bill', hue='sex',estimator=np.var, kind='bar', data=tips)
```

# CountPlot:

```python
sns.catplot(x='sex',hue='smoker', kind='count', data=tips)
```
