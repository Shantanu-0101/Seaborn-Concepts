```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
flights = sns.load_dataset('flights')
```

```python
flights.head()
```

```python
x= flights.pivot_table(index='year', columns='month', values='passengers', aggfunc='sum')
```

```python
sns.heatmap(x, cbar=False)
```

```python
sns.heatmap(x, cbar=False, lw=0.5)
```

```python
sns.heatmap(x, cbar=False, lw=0.5, annot=True, fmt='d')
```

```python
sns.heatmap(x, cbar=False, lw=0.5, annot=True, fmt='d', cmap='viridis')
```

```python
sns.heatmap(x, cbar=False, lw=0.5, annot=True, fmt='d', cmap='summer')
```

```python
sns.heatmap(x, cbar=False, lw=0.5, annot=True, fmt='d', cmap='winter')
```
